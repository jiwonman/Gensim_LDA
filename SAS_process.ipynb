# SAS DATA PreProcessing

```python
# import library

import pandas as pd
```

```python
# takes 5m 
data = pd.read_csv('./data/SAS/SAS_words.csv')

new_df = []
index_list = []
unique = data['DOC_ID'].unique()

for i in unique:
    term = []
    count = []
    process_data = []

    term.append(data[data['DOC_ID'] == i]['Term'].tolist())
    count.append(data[data['DOC_ID'] == i]['count'].tolist())


    for a in range(len(count[0])):
        for b in range(count[0][a]):
            process_data.append(term[0][a])

    refactoring = ', '.join(process_data)
    new_df.append(refactoring)
    index_list.append(i)
    print(f"{i}_unique Start")
```

```python
# list to Series
df_series = pd.Series(new_df, index = index_list)
print(df_series)
```

```python
# Series to DataFrame
preprocess = pd.DataFrame(df_series, columns=['object_name'])
preprocess.index.name="image_id"
print(preprocess)
```

```python
# index sort
preprocess_end = preprocess.sort_values(by='image_id', ascending=True)
preprocess_end
```

```python
# DataFrame to csv
preprocess_end.to_csv('./data/SAS/preprocess_SAS.csv')
```
